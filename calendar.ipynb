```python
import datetime
import numpy as np
import pandas as pd
```

```python
limite = datetime.date(2022,12,31)
inicial = datetime.date(1900,1,1)
fecha = []
fecha.append(inicial)

i=0
while fecha[i]<limite:
    inicial+= datetime.timedelta(days=1)
    fecha.append(inicial)
    i+=1
calendario=pd.DataFrame()
calendario['date']=pd.to_datetime(fecha)
pd.to_datetime(calendario['date'])
calendario['day'] = calendario['date'].dt.day
calendario['month'] = calendario['date'].dt.month
calendario['year'] = calendario['date'].dt.year

calendario
```

```python
calendario['decade']=calendario['year']/10
calendario['decade']=calendario['decade'].apply(np.floor)
calendario['decade']=(calendario['decade']*10).astype(int)
```
